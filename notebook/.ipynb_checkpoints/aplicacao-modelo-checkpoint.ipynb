```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import LeaveOneOut
```

```python
# Carregando Dataset
df = pd.read_csv('data/input-knn.csv',engine='python', sep=',')

df.head()
```

```python
# Separando Alternativas selecionas(X) e conteúdos sugeridos(y)
X, y = df.drop('Content',axis=1).values, df['Content'].values

# Normalizando os id´s das alternativas
scaler = StandardScaler()
X = scaler.fit_transform(X)

X[:1]
```

```python
steps = [10, 20, 30, 40, 50, 60, 70, 80]
train_accuracy_mean, test_accuracy_mean = [], []

for i in steps:
    X_data = X[:i]
    train_accuracy, test_accuracy = np.empty(len(X_data)), np.empty(len(X_data))
    
    loo = LeaveOneOut()
    for train, test in loo.split(X_data):
        X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
        
        knn = KNeighborsClassifier(n_neighbors = 7)
        knn.fit(X_train, y_train)
        
        # Calculando acurácia de treino
        train_accuracy[test] = knn.score(X_train, y_train)
        
        # Calculando acurácia de teste
        test_accuracy[test] = knn.score(X_test, y_test)
    
    train_accuracy_mean.append(train_accuracy.mean() * 100)
    test_accuracy_mean.append(test_accuracy.mean() * 100)
    
    print("tamanho da base: ", i)
    print("acurácia treino: ", train_accuracy.mean() * 100)
    print("acurácia teste: ", test_accuracy.mean() * 100) 
```
