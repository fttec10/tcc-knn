```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import LeaveOneOut
```

```python
# Carregando Dataset
df = pd.read_csv('data/input-knn.csv',engine='python', sep=',')

df.head()
```

```python
# Separando Alternativas selecionas(X) e conteúdos sugeridos(y)
X, y = df.drop('Content',axis=1).values, df['Content'].values

# Normalizando os id´s das alternativas
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Dividindo a base em Treino(80%) e Teste(20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)

print(len(X_train))
print(len(X_test))

#train_accuracy, test_accuracy = np.empty(len(X_data)), np.empty(len(X_data))

knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, y_train)

#print("acurácia treino: ", train_accuracy.mean() * 100)
print("acurácia teste: ", knn.score(X_test, y_test)) 
```

```python
steps = [10, 20, 30, 40, 50, 60, 90]
#train_accuracy_mean, test_accuracy_mean = [], []

test_accuracy = 0

for i in steps:
    X_data = X_train[:i]
    y_data = y_train[:i]
    
    print(len(X_data))
    print(len(X_test))
    
    #train_accuracy, test_accuracy = np.empty(len(X_data)), np.empty(len(X_data))

    knn = KNeighborsClassifier(n_neighbors = 7)
    knn.fit(X_data, y_data)

    # Calculando acurácia de teste
    test_accuracy = knn.score(X_test, y_test)
    
    print("tamanho da base: ", i)
    #print("acurácia treino: ", train_accuracy.mean() * 100)
    print("acurácia teste: ", test_accuracy) 
```
